# Preprocessing in Africa 
 

```python
df_africa = life_expectancy_africa.copy()
```

```python
df_africa = df_africa[['Life expectancy', 'BMI', 'Schooling']].fillna(df_africa[['Life expectancy', 'BMI', 'Schooling']].mean())
```

```python
df_africa[['Life expectancy', 'BMI', 'Schooling']].isnull().sum()
```

```python
# Define bins and labels for BMI categorization
bmi_bins = [df_africa["BMI"].min(), 18.5, 24.9, 29.9, df_africa["BMI"].max()]
bmi_labels = ['Underweight', 'Normal', 'Overweight', 'Obese']

# Create BMI_category
df_africa['BMI_category'] = pd.cut(df_africa['BMI'], bins=bmi_bins, labels=bmi_labels, right=True)

life_bins = [ 0, 60, 70, 80, 100]
life_labels = ['Low', 'Medium', 'High', 'very High']

df_africa['Life_expectancy_category'] = pd.cut(df_africa['Life expectancy'], bins=life_bins, labels=life_labels, right=True)


# Define bins and labels for Schooling categorization
# From describe(): min Schooling is 0.0, max is 20.7
# Using suggested ranges for illustration, can be adjusted based on distribution
schooling_bins = [0, 8, 12, 21]
schooling_labels = ['Low', 'Medium', 'High']

# Create Schooling_category
df_africa['Schooling_category'] = pd.cut(df_africa['Schooling'], bins=schooling_bins, labels=schooling_labels, right=True)

print("BMI Category Value Counts:")
print(df_africa['BMI_category'].value_counts())
print("\nSchooling Category Value Counts:")
print(df_africa['Schooling_category'].value_counts())
print(df_africa['Life_expectancy_category'].value_counts())
```

# Chi 2 

```python
from scipy.stats import chi2_contingency


df_africa['Region'] = 'Africa'
df_western_europe['Region'] = 'Western Europe'

# Combiner les deux régions
df_combined = pd.concat([df_africa, df_western_europe])

# Table de contingence : Region vs BMI_category
contingency_table_bmi = pd.crosstab(df_combined['Region'], df_combined['BMI_category'])
print("Contingency Table (Region vs. BMI_category):")
print(contingency_table_bmi)

# Test Chi2
chi2_bmi, p_bmi, dof, expected = chi2_contingency(contingency_table_bmi)
print(f"\nChi-squared statistic: {chi2_bmi:.2f}")
print(f"P-value: {p_bmi:.4f}")

# Pareil pour Schooling
contingency_table_schooling = pd.crosstab(df_combined['Region'], df_combined['Schooling_category'])
print("\nContingency Table (Region vs. Schooling_category):")
print(contingency_table_schooling)

chi2_schooling, p_schooling, dof, expected = chi2_contingency(contingency_table_schooling)
print(f"\nChi-squared statistic: {chi2_schooling:.2f}")
print(f"P-value: {p_schooling:.4f}")

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Données Afrique
X_africa = df_africa[['BMI', 'Schooling']]
y_africa = df_africa['Life expectancy']

# Séparer en train/test
#X_train, X_test, y_train, y_test = train_test_split(X_africa, y_africa, test_size=0.4, random_state=42)

# Créer et entraîner le modèle
model_africa = LinearRegression()
model_africa.fit(X_africa, y_africa)  # Entraîner sur TRAIN

# Prédictions
#y_pred = model_africa.predict(X_test)  # Prédire sur TEST

# Résultats
print("Coefficients:", model_africa.coef_)
print("Intercept:", model_africa.intercept_)
print("R²:", model_africa.score(X_africa, y_africa))  # Score sur TEST
```
